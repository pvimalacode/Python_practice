```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

used seed to get same result after every random generation

```python
np.random.seed(0)
df1 = pd. DataFrame(-2*np.random.rand(50,1), columns=['x'])
np.random.seed(1)
df1['y'] = 1 +  np.random.rand(50,1)*2
df1.head()
```

```python
df1.info()
```

```python
import seaborn as sns
sns.scatterplot(data=df1,x='x',y='y')
```

<h2><b>  ELBOW METHOD: calculated the errors for clusters size k = 1 to 11 to find the best size with minimum error/distortion

```python
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(df1)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.grid()
plt.show()
```

<h2><b><u>CONCLUSION:</u><br>•	From the above we can conclude that, error is almost constant after 6 clusters. <br>
•	There is not much significant improvement in error when we go further clustering greater than 6.<br>
•	It is not advisable to have have too many clusters. So we will with number of clusters as 6 <br>
•	We are not increasing it further as it will lead to overfitting.<br>
•	We plotted the visualization of the clustering as well, along with its centroids as per number of clusters=7.
<br>


```python
from sklearn.cluster import KMeans

km = KMeans(
    n_clusters=7, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0
)
y_km = km.fit_predict(df1)
```

```python
centroids = km.cluster_centers_
```

```python
centroids
```

```python
plt.scatter(df1['x'], df1['y'], c= km.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```
