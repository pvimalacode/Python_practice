<h2><b> 3. Applying Stadardisation Techniques


Used Robust scalings as there are many outliers

```python
from sklearn.preprocessing import RobustScaler
trans= RobustScaler()
data=trans.fit_transform(wine_trans)
```

```python
wine_stand= pd.DataFrame(data,columns=wine_df.iloc[:,1:-2].columns)
```

<h3><b>The Data frame after applying the standdardization on all numerical non - categorical features

```python
wine_stand.head()
```

<h3><b>Added the categorical variables to the standardised data

```python
wine_stand['type']=wine_df['type'].values
wine_stand['quality']=wine_df['quality'].values
wine_stand.head()
```
