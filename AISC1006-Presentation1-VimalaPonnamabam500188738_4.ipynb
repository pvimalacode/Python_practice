```python
#reading all the files for analysis

amazon=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Amazon.csv')
apple=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Apple.csv')
facebook=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Facebook.csv')
google=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Google.csv')
netflix=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Netflix.csv')

prefix=['facebook_','amazon_','apple_','netflix_','google_']
names=[facebook,amazon,apple,netflix,google]


facebook['company']='Facebook'
apple['company']='Apple'
amazon['company']='Amazon'
netflix['company']='Netflix'
google['company']='Google'
```

Adding Prefix

```python
for count in range(len(names)):
  names[count]=pd.DataFrame(names[count]).add_prefix(prefix[count])
```




*   converting Date column to Datetime format
*   Renaming the Date column
*   Removing NaN values
*   Renaming the Date column










```python
for name in names:
  name[name.columns[0]]=name[name.columns[0]].apply(lambda x : pd.to_datetime(x))
  name.rename(columns={name.columns[0]:'Date'},inplace=True)
  name=name.replace(np.nan,0)
```

```python
facebook=names[0]
amazon=names[1]
apple=names[2]
netflix=names[3]
google=names[4]
```

Merging files on Date field

```python
final=pd.merge(apple,amazon,how='outer' ,on='Date') 
final1=pd.merge(final,netflix,how='outer' ,on='Date') 
final2=pd.merge(final1,google,how='outer' ,on='Date') 
ds_merge=pd.merge(final2,facebook,how='outer' ,on='Date') 
```

```python
ds_merge['Date']=ds_merge['Date'].apply(lambda x:x.year)
ds_merge1=copy.deepcopy(ds_merge)
ds_merge.replace(np.nan,0,inplace=True)
```

```python
ds_merge1.head()
```

```python
ds_merge1=ds_merge[['Date','facebook_Close','amazon_Close','apple_Close','netflix_Close','google_Close']]
ds_bcr=ds_merge[['Date','facebook_Volume','amazon_Volume','apple_Volume','netflix_Volume','google_Volume']]
```

```python
ds_merge1=ds_merge1.groupby('Date').mean()
ds_bcr=ds_bcr.groupby('Date').mean()
```
