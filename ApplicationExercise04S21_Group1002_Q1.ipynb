```python

#print(f.read())
#import library to validate input variable
import pyinputplus as pyip

    
#define display of menu function
def dis():
    print('Contact Manager\nCOMMAND MENU\nlist - Display all contacts\nview - View a contact\nadd - Add a contact\ndel - Delete a contact\nexit - Exit program')

    
    #return(cmd)

#function to read and return the details as list
def r_file():
    file = open("C:\DS&AI\Vimala\Python\Python_Class\Pythongrp4\input.txt", "r+",newline=None)
    with open('input.txt') as file:
        members=[]
        for line in file:
            line=line.replace('\n','')
            members.append(line.strip())
            members=[i for i in members if i!='']
    file.close()
    return(members)

#function to display all the name
def listing():
    members=[]
    members=r_file() # read the file
    #print(members)
    for item in range(int((len(members)+1)/3)): # loop through to get the names
        #print(members[item*3])
        print(item+1,members[(item)*3])

#function to get all details from the serial number of name
def v_contact():
    members=r_file()
    itm_nbr=pyip.inputNum('Number:',min=1,max=int((len(members)/3))) # check to enter valid number
    #itm_nbr=itm_nbr-1
    #print(members)
    print()
    print(*members[(itm_nbr-1)*3:itm_nbr*3],sep='\n')

#function to add name,email,phone details to file and save
def add_det():
    name=input('Name:')
    email=input('Email:')
    phone=(input('Phone:'))
    with open('input.txt','a') as file:
        #file.write('\n')
        file.write('\n'+name+'\n'+email+'\n'+phone)
    #file.close()
    print(name,'was added')
    
#function to delete an entry from file
def del_data():
    l=r_file() 
    # store data from file
    del_num=pyip.inputNum('Enter item to delete:',min=1,max=int((len(l)/3))) 
    # validation of input 
    with open('input.txt','w',newline=None) as file:
        file.truncate() 
        # delete all
    print(l[(del_num-1)*3].replace('\n',''),' is deleted',sep='') 
    # remove empty and extra next lines
    for i in range(3): 
        # delete the entered serial number entry
        #print(i)
        l.pop((del_num-1)*3)

    #write in the file the data
    with open('input.txt','a',newline='\n') as file:
        #print(len(l))
        for word in range(len(l)):
            if word==len(l)-1:
                file.write(l[word])
                #print(word)
            else:    
                file.write(l[word]+'\n')
                #print(word)
    file.close()

    
#function to get command
def getcmd():
    cmd=input('Command:')
```

```python
cmd='start'
dis()
while cmd!='exit':
    cmd=input('Command:')
    
    if cmd=='list':
        listing()
        
    elif cmd=='view':
        v_contact()
    elif cmd=='add':
        add_det()
    elif cmd=='del':
        del_data()
    elif cmd=='exit':
        print('Bye!')
    else:
        print("enter valid inputs only as per below\nlist\nview\nadd\ndel\nexit")
#print('Bye!')
        
```
