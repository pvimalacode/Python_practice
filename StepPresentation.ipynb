Adding required files for analysis

```python
#reading all the files for analysis

amazon=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Amazon.csv')
apple=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Apple.csv')
facebook=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Facebook.csv')
google=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Google.csv')
netflix=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Netflix.csv')
```

Adding Prefix to files

```python
files=[facebook,amazon,apple,google,netflix]
prefix=['facebook_','amazon_','apple_','netflix_','google_']
names=[facebook,amazon,apple,netflix,google]
```

```python
facebook['company']='Facebook'
apple['company']='Apple'
amazon['company']='Amazon'
netflix['company']='Netflix'
google['company']='Google'
```

```python
apple.isna().sum()
```

```python
#print(apple.isna().sum())
apple=apple.dropna(how='any')
apple.isna().sum()
```

Adding Prefix

```python
for count in range(len(files)):
  names[count]=pd.DataFrame(files[count]).add_prefix(prefix[count])
  
```

converting Date column to Datetime format

```python
for name in names:
  name[name.columns[0]]=name[name.columns[0]].apply(lambda x : pd.to_datetime(x))
```

```python
#final['Volume']=final['Volume'].apply(lambda x: np.log(x) if x!=0 else 0)
```

Renaming the Date column

```python
for name in names:
  name.rename(columns={name.columns[0]:'Date'},inplace=True)
```

Removing NaN values

```python
for name in names:
  name=name.replace(np.nan,0)
```

```python
facebook=names[0]
amazon=names[1]
apple=names[2]
netflix=names[3]
google=names[4]
```

Merging files on Date field

```python
final=pd.merge(facebook,amazon,how='outer' ,on='Date') 
final1=pd.merge(final,apple,how='outer' ,on='Date') 
final2=pd.merge(final1,netflix,how='outer' ,on='Date') 
final3=pd.merge(final2,google,how='outer' ,on='Date') 
```

```python
final3['Date']=final3['Date'].apply(lambda x:x.year)
```

Removing NaN values after merging

```python
final3.replace(np.nan,0,inplace=True)
final3.shape
```

DataFrame after merging

```python
final3.head()
```

Grouping The data frame by Date field

```python
final3=final3.groupby('Date').median()
```
